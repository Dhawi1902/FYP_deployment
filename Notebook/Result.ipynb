```python
import json
import matplotlib.image as mpimg
import os
import matplotlib.pyplot as plt
```

```python
# Task 1: Import images with labels and print in a tabular format
data_folder = "/home/dhawi/Documents/dataset"
dataset = data_folder + "/AI_project"
model_folder = "/home/dhawi/Documents/model"
history_folder = "/home/dhawi/Documents/History"
```

```python
def plot_training_history(history):
    plt.figure(figsize=(12, 4))

    # Plot training & validation accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Val Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot training & validation loss
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()
```

```python
def show_history(model_name, gan_history, gan_type):
    plt.figure(figsize=(6, 4))

    # Plot training & validation accuracy
    plt.plot(gan_history['Generator Loss'], label='Generator Loss')
    plt.plot(gan_history['Discriminator Loss'], label='Discriminator Loss')
    plt.title(model_name + ' ' + gan_type + ' loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend()
    plt.savefig(os.path.join('static/images/'+ gan_type + '_' + model_name + '.png'))
    plt.show()
```

```python
def show_vae_history(model_name, history):
    plt.figure(figsize=(6, 4))
    plt.plot(history['loss'], label='loss')
    plt.plot(history['reconstruction_loss'], label='reconstruction Loss')
    plt.plot(history['kl_loss'], label='kl Loss')
    plt.title(model_name + ' vae loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend()
    plt.savefig(os.path.join('static/images/vae_' + model_name + '.png'))    
    plt.show()
```

```python
classes = ['caries', 'gingivitis', 'wsl']
```

```python
for model_name in classes:      
    gan_history_path = history_folder + "/fyp_" + model_name + "_GAN1024_history.json"
    edgan_history_path = history_folder + "/fyp_" + model_name + "_EDGAN1024_history.json"
    history_path = history_folder + "/" + model_name + "_vae1024_history.json"
    history = json.load(open(history_path, 'r'))
    gan_history = json.load(open(gan_history_path, 'r'))
    edgan_history = json.load(open(edgan_history_path, 'r'))
    show_history(model_name, gan_history, 'GAN')
    show_history(model_name, edgan_history, 'EDGAN')
    show_vae_history(model_name, history)
```
